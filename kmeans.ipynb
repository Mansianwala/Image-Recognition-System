```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from skimage.feature import hog
from keras.applications import VGG16
from keras.datasets import cifar10
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from keras.models import Model, Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout, BatchNormalization
from keras.optimizers import Adam
from keras.preprocessing.image import ImageDataGenerator
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()


# Preprocess images
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255


# Extract HOG features
def extract_hog_features(images):
    hog_features = []
    for image in images:
        hog_feat = hog(image, pixels_per_cell=(8, 8), cells_per_block=(2, 2), multichannel=True)
        hog_features.append(hog_feat)
    return np.array(hog_features)


# Extract CNN features
def extract_cnn_features(images):
    base_model = VGG16(weights='imagenet', include_top=False)
    model = Model(inputs=base_model.input, outputs=base_model.get_layer('block5_conv2').output)
    features = model.predict(images)
    return features.reshape(features.shape[0], -1)

# Combine HOG and CNN features
def combine_features(hog_features, cnn_features):
    return np.hstack((hog_features, cnn_features))


# Apply K-means clustering
# Apply K-means clustering
def apply_kmeans(features, n_clusters=10):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(features.astype(np.float32))  # Convert to float32
    return kmeans


def retrieve_similar_images(query_index, features, kmeans, x_train, y_train, n=10):
    query_image = x_train[query_index]
    hog_feat = extract_hog_features(np.expand_dims(query_image, axis=0))
    cnn_feat = extract_cnn_features(np.expand_dims(query_image, axis=0))
    query_features = combine_features(hog_feat, cnn_feat)
    print("Query features shape:", query_features.shape)
    cluster_label = kmeans.predict(query_features)
    cluster_indices = np.where(kmeans.labels_ == cluster_label)[0]
    distances = np.linalg.norm(features[cluster_indices] - query_features, axis=1)
    sorted_indices = cluster_indices[np.argsort(distances)]
    similar_indices = sorted_indices[:n]
    similar_labels = y_train[similar_indices]
    return similar_indices, similar_labels

def compute_retrieval_accuracy(query_index, features, kmeans, x_train, y_train):
    true_label = y_train[query_index]
    similar_indices, similar_labels = retrieve_similar_images(query_index, features, kmeans, x_train, y_train)
    correct = np.sum(similar_labels == true_label)
    accuracy = correct / len(similar_labels)
    return accuracy


# Training phase
hog_features_train = extract_hog_features(x_train)
cnn_features_train = extract_cnn_features(x_train)
combined_features_train = combine_features(hog_features_train, cnn_features_train)
print("Comined features dimension:", combined_features_train.shape)
kmeans = apply_kmeans(combined_features_train)

# Calculate training accuracy
train_accuracy = model.evaluate(x_train, y_train)[1]
print("Training Accuracy:", train_accuracy)

# Testing phase
hog_features_test = extract_hog_features(x_test)
cnn_features_test = extract_cnn_features(x_test)
hog_features_test = hog_features_test.astype(combined_features_train.dtype)
cnn_features_test = cnn_features_test.astype(combined_features_train.dtype)
combined_features_test = combine_features(hog_features_test, cnn_features_test)

# Convert the data type of training features to double
combined_features_train = combined_features_train.astype(np.double)
```

```python

# Apply K-means clustering to training features
# Apply K-means clustering to training features
kmeans = apply_kmeans(combined_features_train)

# Convert the data type of testing features to float32
combined_features_test = combined_features_test.astype(np.float32)

# Calculate testing accuracy
test_labels = kmeans.predict(combined_features_test)
test_accuracy = accuracy_score(y_test, test_labels)
print("Testing Accuracy:", test_accuracy)


# Output similar images
# Example usage
query_index = int(input("Enter the index of the query image: "))  # Index of the query image

retrieval_accuracy = compute_retrieval_accuracy(query_index, combined_features_train, kmeans, x_train, y_train)
print("Retrieval Accuracy:", retrieval_accuracy)



similar_indices, similar_labels = retrieve_similar_images(query_index, combined_features_train, kmeans, x_train, y_train)

# Display similar images
plt.figure(figsize=(10, 5))
for i, idx in enumerate(similar_indices):
    plt.subplot(1, len(similar_indices), i + 1)
    plt.imshow(x_train[idx])
    plt.axis('off')
plt.show()



# Get predicted probabilities for each class
train_probabilities = model.predict(x_train)

# Convert probabilities to class labels
train_predictions = np.argmax(train_probabilities, axis=1)

# Plot confusion matrix for training data
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.heatmap(confusion_matrix(y_train, train_predictions), annot=True, fmt="d", cmap="Blues")
plt.title("Training Data Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")

# Repeat the same process for testing data
test_probabilities = model.predict(x_test)
test_predictions = np.argmax(test_probabilities, axis=1)

plt.subplot(1, 2, 2)
sns.heatmap(confusion_matrix(y_test, test_predictions), annot=True, fmt="d", cmap="Blues")
plt.title("Testing Data Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")

plt.show()

```
